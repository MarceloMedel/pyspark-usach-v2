```python
import socket
import time
import json

def start_server(file_path, host, port, delay):
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as server_socket:
        server_socket.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
        server_socket.bind((host, port))
        server_socket.listen(1)
        print(f"Server listening on {host}:{port}")
        conn, addr = server_socket.accept()
        with conn:
            print(f"Connected by {addr}") 
            with open(file_path, 'r') as file: 
                json_file = json.load(file) 
                for line in json_file:
                    conn.sendall((json.dumps(line) + '\n').encode())
                    print(f"registro enviado: {line}") 
                    time.sleep(delay) 
            print("File data sent, closing connection.")
            conn.close()

host = 'localhost'
port = 9999
delay = 1
file_path = 'data/streaming/streaming_sample.json'

start_server(file_path, host, port, delay)
```
